```python
import requests
from config import pro_api_key
from config import congress_api_key
from pprint import pprint
import json
```

```python
# Try call to congress.gov API for all the bills in 2022 for the 117th Congress:
# query = ""
url = "https://api.congress.gov/v3/bill/117?format=json&offset=0&limit=250&fromDateTime=2022-01-01T00%3A00%3A00Z&toDateTime=2022-12-13T00%3A00%3A00Z&sort=updateDate%2Basc"
# url = f"https://api.propublica.org/congress/v1/bills/search.json?query={query}"
response = requests.get(url, headers={'X-API-Key':congress_api_key})
pprint(response.text)

with open('test.json','w') as file:
    json.dump(response.text, file)
```
